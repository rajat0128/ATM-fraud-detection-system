```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os, warnings
warnings.filterwarnings('ignore')

# Create project folders
for folder in ['../data','../models','../reports','../app']:
    os.makedirs(folder, exist_ok=True)
print("Project folders ready!")
```

```
# Load the ULB Credit Card dataset
df = pd.read_csv('../data/creditcard.csv')
print("Dataset loaded!")
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```
# First look at the data
print("=== First 5 rows ===")
print(df.head())
print("\n=== Column types ===")
print(df.dtypes)
print("\n=== Basic statistics ===")
print(df.describe().round(3))
```

```
# Check for null values
print("=== Null values per column ===")
print(df.isnull().sum())
print("\nTotal nulls:", df.isnull().sum().sum())
```

```
# Class distribution — the imbalance problem
print("=== Fraud vs Normal ===")
print(df['Class'].value_counts())
print(f"\nFraud percentage: {round(df['Class'].mean()*100, 4)}%")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
counts = df['Class'].value_counts()
colors = ['steelblue','crimson']

axes[0].bar(['Normal','Fraud'], counts.values, color=colors, edgecolor='black')
axes[0].set_title('Transaction Count')
axes[0].set_ylabel('Count')
for i,v in enumerate(counts.values):
    axes[0].text(i, v+500, str(v), ha='center', fontweight='bold')

axes[1].pie(counts.values, labels=['Normal','Fraud'],
            colors=colors, autopct='%1.3f%%', startangle=90)
axes[1].set_title('Percentage Split')

plt.tight_layout()
plt.savefig('../reports/phase1_class_distribution.png', dpi=150)
plt.show()
print("Chart saved to reports/!")
```

```
# Amount distribution
print("Amount statistics:")
print(df['Amount'].describe().round(2))
print(f"\nMax amount : {df['Amount'].max()}")
print(f"Mean amount: {df['Amount'].mean():.2f}")
print("\nPhase 1 complete! Dataset understood.")
```
